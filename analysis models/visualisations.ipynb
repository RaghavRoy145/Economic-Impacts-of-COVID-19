#  GDP 

```python
barWidth = 0.15
fig = plt.subplots(figsize =(12, 8)) 
IT = [228.75532, 163.58355946, 166.95715873]
ECE=[218.9773162, 128.62094,156.7372877]
br1 = np.arange(3) 
br2 = [x + barWidth for x in br1] 
   
# Make the plot 
plt.bar(br1, IT, color ='b', width = barWidth, 
        edgecolor ='grey', label ='IT') 
plt.bar(br2, ECE, color ='g', width = barWidth, 
        edgecolor ='grey', label ='ECE')  
# Adding Xticks  
plt.xlabel('Countries', fontweight ='bold') 
plt.ylabel('GDP', fontweight ='bold') 
plt.xticks([r + barWidth for r in range(len(IT))], 
           ['USA', 'IND', 'ZAF']) 
plt.legend(labels=['predicted', 'Actual'])
plt.show() 
```
